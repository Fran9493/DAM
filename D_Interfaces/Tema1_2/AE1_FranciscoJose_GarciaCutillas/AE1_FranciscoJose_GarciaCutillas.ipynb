## Ejercicio 3

Define una función llamada is_romano() dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano.

```python
import re
"""
Clase Numero. Al crear un objeto de la misma introduciéndole un valor entero, nos permite 
obtener el valor en entero y en número romano de dicho valor.
"""
class Numero:
    
    #Constructor de la clase
    def __init__(self, numero):

        self.numeroEntero = numero
        self.numeroRomano = self.convierteRomano(numero)
    
    #Método que recibe como parámetro un número entero y devuelve su equivalente en número romano
    def convierteRomano(self,numEntero):

        millares = "M"
        centenas = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
        decenas = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
        unidades = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]
        
        resMillares = millares * (int(numEntero / 1000)) 
        resCentenas = centenas[int((numEntero % 1000) / 100)]
        resDecenas = decenas[int((numEntero % 100) / 10)]
        resUnidades = unidades[int(numEntero % 10)]
        
        resultado = resMillares + resCentenas + resDecenas + resUnidades
        
        return resultado 
    
    #Método que muestra por terminal la equivalencia del número entero del objeto con su equivalente en romano
    def imprime(self):

        print(" La equivalencia de " + str(self.numeroEntero) + " en número romano es: " + str(self.numeroRomano))
    
    #Método para sumar dos números romanos
    def suma_romano(self,numRomanoSuma):

        romanoEntero = self.convierteEntero(numRomanoSuma)
        self.numeroEntero = self.numeroEntero + romanoEntero
        self.numeroRomano = self.convierteRomano(self.numeroEntero)
              
    #Método para convertir un número romano a entero
    def convierteEntero(self, numRomano):        
        numEntero = 0
    
        # Diccionario que contiene el carácter en romano y su equivalencia en entero
        romanoEntero = {'M': 1000, 'D': 500, 'C': 100, 'L': 50, 'X': 10, 'V': 5, 'I': 1}

        for i in range(len(numRomano)):
            
            valorActual = romanoEntero[numRomano[i]]
           
            if i < len(numRomano) - 1 and romanoEntero[numRomano[i + 1]] > valorActual:
                numEntero -= valorActual
            else:
                numEntero += valorActual

        return numEntero   

    #Método is_romano() que devuelve true si la cadena cumple la nomenclatura de un número romano, false en el caso contrario   
    def is_romano(self, cadena):

        patron = "^(?=[MDCLXVI])M*(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$"
        if(re.search(patron, cadena)):
            return True
        else:
            raise ValueError("Error, la cadena introducida no se corresponde a un número romano")
    
```

Implementa una clase llamada MejorNumero. Esta clase heredará las propiedades de Numero e incluirá dos métodos nuevos, llamados resta() y multiplica(), para restar y multiplicar los atributos recibiendo por parámetro otro número romano. 


## Ejercicio 5

En la clase MejorNumero, crea un nuevo método llamado iterar() que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). Los posibles errores se tendrán que gestionar con excepciones (try...except...) para mostrar un mensaje y seguir ejecutando el siguiente número.

**NOTA:** el mensaje de error ha de seguir el siguiente formato --> print('Ha fallado el número', 5). 

```python
"""
Clase MejorNumero que hereda de Numero. Incluye dos nuevos métodos para restar y multiplicar los atributos recibiendo por 
parámetro otro número romano
"""
class MejorNumero(Numero):

    #Constructor de la clase
    def __init__(self, numero):

        self.numeroEntero = numero
        self.numeroRomano = self.convierteRomano(numero)

    #Método resta(). Resta a los atributos el valor romano recibido por parámetro
    def resta(self, numRomanoRestar):

        numClaseInt = self.convierteEntero(self.numeroRomano)
        numRecibidoInt = self.convierteEntero(numRomanoRestar)

        resta = numClaseInt - numRecibidoInt

        self.numeroRomano = self.convierteRomano(resta)
        self.numeroEntero = resta
    
    #Método multiplica(). Multiplica a los atributos el valor romano recibido por parámetro
    def multiplica(self, numRomanoMultiplicar):

        numClaseInt = self.convierteEntero(self.numeroRomano)
        numRecibidoInt = self.convierteEntero(numRomanoMultiplicar)

        multiplicacion = numClaseInt * numRecibidoInt

        self.numeroRomano = self.convierteRomano(multiplicacion)
        self.numeroEntero = multiplicacion

    #Método iterar(). Recibe una lista de 3 números romanos y los va a ir sumando a los atributos de la clase
    def iterar(self, listaRomanos):
        
        for elemento in listaRomanos:

            try:

                self.suma_romano(elemento)

            except TypeError:

                print("Ha fallado el número " + str(elemento))
    
```

```python
#Pruebas método resta() de la clase MejorNumero que hereda de Numero
print("\n** Pruebas del método resta() **")
num = MejorNumero(45)
num.resta('X')
print("El resultado de restar X a 45 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))
num = MejorNumero(120)
num.resta('IX')
print("El resultado de restar IX a 120 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))
num = MejorNumero(566)
num.resta('D')
print("El resultado de restar D a 566 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))

#Pruebas método multiplica() de la clase MejorNumero que hereda de Numero
print("\n** Pruebas del método multiplica() **")
num = MejorNumero(25)
num.multiplica('V')
print("El resultado de multiplicar V a 25 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))
num = MejorNumero(5)
num.multiplica('XV')
print("El resultado de multiplicar XV a 5 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))
num = MejorNumero(200)
num.multiplica('III')
print("El resultado de multiplicar III a 200 en romano es " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))

#Pruebas método iterar() de la clase MejorNumero que hereda de Numero
print("\n** Pruebas del método iterar() **")
listaRomanos = ['V', 'XX', 'II']
num = MejorNumero(10)
num.iterar(listaRomanos)
print("El resultado de iterar V, XX, II al número 10 " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))

listaRomanos1 = ['IV', 'C', 'D']
num = MejorNumero(50)
num.iterar(listaRomanos1)
print("El resultado de iterar IV, C, D al número 50 " + str(num.numeroRomano) + " y en número entero es " + str(num.numeroEntero))

```
